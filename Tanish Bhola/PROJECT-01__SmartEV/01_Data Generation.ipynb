```python
import pandas as pd
import numpy as np
```

```python
data = []
```

```python
LeftLow = 7
LeftHigh = 7.5

midLeftLow = 7.5
midLeftHigh = 8

midRightLow = 8
midRightHigh = 8.5

RightLow = 8.5 
RightHigh = 9

for i in range(1,21):
    
    
    #AM = 0
    #PM =1
    if i%2 == 0:
        # Time : 3,9
        data.append([i,3,0,round(np.random.uniform(LeftLow,LeftHigh),2)])
        data.append([i,9,0,round(np.random.uniform(RightLow,RightHigh),2)])
        data.append([i,9,1,round(np.random.uniform(midRightLow,midRightHigh),2)])
        data.append([i,3,1,round(np.random.uniform(midLeftLow,midLeftHigh),2)])
        if i<6:
            LeftLow += 0.5
            LeftHigh += 0.5
            RightLow += 0.5
            RightHigh += 0.5
            midLeftLow += 0.5
            midLeftHigh += 0.5
            midRightLow += 0.5
            midRightHigh += 0.5



    else:
        #Time : 6,12
        data.append([i,6,0,round(np.random.uniform(LeftLow,LeftHigh),2)])
        data.append([i,12,0,round(np.random.uniform(midLeftLow,midLeftHigh),2)])
        data.append([i,12,1,round(np.random.uniform(midRightLow,midRightHigh),2)])
        data.append([i,6,1,round(np.random.uniform(RightLow,RightHigh),2)])
            

```

```python
len(data)
```

```python
dataPanda = pd.DataFrame(data , columns=['numberOfCars','time','dayNight','Price'])
```
