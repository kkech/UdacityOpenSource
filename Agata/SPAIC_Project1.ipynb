```
def get_paraller_db(db, remove_index):
  

  return torch.cat((db[0:remove_index], 
                    db[remove_index+1:]))
```

```
def get_parallel_dbs(db):
  
  parallel_dbs = list()

  for i in range(len(db)):
     pdb = get_paraller_db(db, i)
     parallel_dbs.append(pdb)
    
  return parallel_dbs
```
